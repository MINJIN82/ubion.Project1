```python
import pandas as pd
```

## 환율, IRS

```python
# 데이터 불러오기
irs = pd.read_excel('./엑셀/찐_IRS_10.xlsx')
ex = pd.read_excel('./엑셀/환율_10.xlsx')
```

```python
# 개수 확인
irs
```

```python
# 개수 확인
ex
```

```python
# 환율 '일자' 칼럼명을 'Datetime'으로 변경 (for merge)
ex = ex.rename(columns={'일자':'DateTime'})
ex
```

```python
# 비교 위해 두 컬럼씩만 가져오기
irs = irs[["DateTime", "Mid"]]
ex = ex[["DateTime", "종가"]]
ex
```

```python
# 둘 다 datetime 인지 확인
ex.info()
```

```python
# left merge 후 결측치 확인
irs_ex = pd.merge(irs, ex, on="DateTime", how='left')
irs_ex.isna().sum()
```

```python
# nan값 있는 행 추출
irs_ex[irs_ex['종가'].isnull()]
```

## NDF, 환율

```python
# 엑셀, ndf 파일 불러오기
ex = pd.read_excel('./엑셀/환율_10.xlsx', index_col=0)
ndf = pd.read_excel('./엑셀/NDF_10.xlsx', index_col = 0)
ndf
```

```python
# ndf, ex의 '일자' 칼럼명을 'Datetime'으로 변경 (for merge)
ndf = ndf.rename(columns={'일자':'DateTime'})
ex = ex.rename(columns={'일자':'DateTime'})
ex
```

```python
# 환율 칼럼을 두 컬럼만 가져오기(보기 쉽게 하기위해)
ex = ex[["DateTime", "종가"]]
```

```python
# 둘 다 datetime 인지 확인
ndf.info()
```

```python
# right merge 후 결측치 확인 (ndf 기준으로 merge)
ex_ndf = pd.merge(ex, ndf, on="DateTime", how='right')
ex_ndf.isna().sum()
```

```python
# 엑셀 파일로 만들기
ex_ndf.to_excel('./NDF_EX_10.xlsx')
```
