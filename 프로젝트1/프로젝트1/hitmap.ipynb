```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../데이터자료/join_data.xlsx")
```

```python
df.info()
```

```python
df["M1_스왑포인트"] = df["M1_스왑포인트"].astype(float64)
```

```python
df['1Y_전일비_irs'] = df['1Y_Mid_irs']-df['1Y_Mid_irs'].shift(1)
df['2Y_전일비_irs'] = df['2Y_Mid_irs']-df['2Y_Mid_irs'].shift(1)
df['3Y_전일비_irs'] = df['3Y_Mid_irs']-df['3Y_Mid_irs'].shift(1)
df['5Y_전일비_irs'] = df['5Y_Mid_irs']-df['5Y_Mid_irs'].shift(1)
df['10Y_전일비_irs'] = df['10Y_Mid_irs']-df['10Y_Mid_irs'].shift(1)
df['1Y_전일비_crs'] = df['1Y_Mid_crs']-df['1Y_Mid_crs'].shift(1)
df['2Y_전일비_crs'] = df['2Y_Mid_crs']-df['2Y_Mid_crs'].shift(1)
df['3Y_전일비_irs'] = df['3Y_Mid_irs']-df['3Y_Mid_crs'].shift(1)
df['5Y_전일비_irs'] = df['5Y_Mid_irs']-df['5Y_Mid_crs'].shift(1)
df['10Y_전일비_irs'] = df['10Y_Mid_irs']-df['10Y_Mid_crs'].shift(1)
```

```python
df['국고1년_전일비'] = df['국고1년']-df['국고1년'].shift(1)
df['국고3년_전일비'] = df['국고3년']-df['국고3년'].shift(1)
df['국고5년_전일비'] = df['국고5년']-df['국고5년'].shift(1)
df['국고10년_전일비'] = df['국고10년']-df['국고10년'].shift(1)
df['통안364일_전일비'] = df['통안364일']-df['통안364일'].shift(1)
df['통안2년_전일비'] = df['통안2년']-df['통안2년'].shift(1)
```

```python
df['1Y_베이시스_전일비'] =df['1Y_베이시스'] - df['1Y_베이시스'].shift(1)
df['2Y_베이시스_전일비'] =df['2Y_베이시스'] - df['2Y_베이시스'].shift(1)
df['3Y_베이시스_전일비'] =df['3Y_베이시스'] - df['3Y_베이시스'].shift(1)
df['5Y_베이시스_전일비'] =df['5Y_베이시스'] - df['5Y_베이시스'].shift(1)
df['10Y_베이시스_전일비'] =df['10Y_베이시스'] - df['10Y_베이시스'].shift(1)
```

```python
df = df.drop(['Unnamed: 0','종가_ex','1Y_Mid_irs','2Y_Mid_irs','국고1년', '국고3년', '국고5년', '국고10년'], axis = 1)
df = df.drop(['3Y_Mid_irs','5Y_Mid_irs','10Y_Mid_irs','1Y_Mid_crs','2Y_Mid_crs'], axis = 1)
df = df.drop(['3Y_Mid_crs','5Y_Mid_crs', '10Y_Mid_crs', 'Bid_ndf', 'Ask_ndf','Mid_ndf',], axis = 1)
df = df.drop(['통안364일', '통안2년',  '1Y_베이시스','증감률_ex', '2Y_베이시스', '3Y_베이시스', '5Y_베이시스','10Y_베이시스'], axis = 1)
```
