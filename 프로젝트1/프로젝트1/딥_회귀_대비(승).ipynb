```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')

# 파일 불러오기
all= pd.read_excel("./xlsx/시차상관분석6Data.xlsx", index_col = 0)              # index_col = 0 파일 읽어올 떄 "Unnamed: 0" 컬럼이 추가돼서 해당 컬럼 제거

# # 컬럼 추출
# all = all['DateTime', '종가_ex','국고1년', '국고3년', '국고5년',
#        '국고10년', '통안364일', '통안2년', 'Mid_ndf',
#        '1Y_베이시스', '2Y_베이시스', '3Y_베이시스', '5Y_베이시스', '10Y_베이시스', 'M1_스왑포인트']]            # [[]] 대괄호 2개 사용 -> 데이터 프레임형태로 나옴

# 대비 계산
all['대비_국고1년'] = all['국고1년']-all['국고1년'].shift(1)
all['대비_국고3년'] = all['국고3년']-all['국고3년'].shift(1)
all['대비_국고5년'] = all['국고5년']-all['국고5년'].shift(1)
all['대비_국고10년'] = all['국고10년']-all['국고10년'].shift(1)
all['대비_통안1년'] = all['통안364일']-all['통안364일'].shift(1)
all['대비_통안2년'] = all['통안364일']-all['통안364일'].shift(1)
all['대비_베이시스1년'] = all['1Y_베이시스']-all['1Y_베이시스'].shift(1)
all['대비_베이시스2년'] = all['2Y_베이시스']-all['2Y_베이시스'].shift(1)
all['대비_베이시스3년'] = all['3Y_베이시스']-all['3Y_베이시스'].shift(1)
all['대비_베이시스5년'] =all['5Y_베이시스']-all['5Y_베이시스'].shift(1)
all['대비_베이시스10년'] = all['10Y_베이시스']-all['10Y_베이시스'].shift(1)
all['대비_ndf'] = all['Mid_ndf']-all['Mid_ndf'].shift(1)
all['스왑포인트_1월물'] = all["M1_스왑포인트"]/100 

# 전일종가 계산
all['전일종가_ex'] = all['종가_ex'].shift(1)

# Nan 행 지우고 인덱스 초기화
all.drop([0], axis=0, inplace=True)
all.reset_index(drop=True, inplace=True)

# # 필요 칼럼만 남기기
# all = all.copy()
# all = all[["DateTime", "종가_ex", "전일종가_ex", "대비_국고1년", "대비_국고3년", "대비_국고5년", "대비_국고10년", "대비_통안1년", "대비_통안2년", "대비_베이시스1년", "대비_베이시스2년", "대비_베이시스3년", "대비_베이시스5년","대비_베이시스10년","대비_ndf", "스왑포인트_1월물"]]

# # DateTime을 인덱스로 바꿔주기
# all = all.set_index('DateTime')
# all
```

```python
x = all[['1Y_전일비_irs', '3Y_전일비_irs',
          '10Y_전일비_irs', '1Y_전일비_crs', '3Y_전일비_crs', 
          '국고3년대비', '국고5년대비', '국고10년대비',
          '통안1년대비','M1_스왑포인트', '전날 종가_ex', '종가_NDF차이']]
y = all["종가_ex"]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
import tensorflow as tf 

tf.random.set_seed(3)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)

model = Sequential()
model.add(Dense(10, input_dim = 12, activation = 'linear'))
model.add(Dense(5, activation = 'linear'))
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

history = model.fit(x, y, epochs=70)
```
