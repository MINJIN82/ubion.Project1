## 1Y, 2Y, 3Y, 5Y, 10Y

```python
import pandas as pd
import openpyxl
import os

file_list=[]
path = './SwapBasis/1Y/'
file_path = os.listdir(path)
for i in file_path:
    file_list.append(i)
    file_list.sort()
file_list
```

```python
y1 = pd.DataFrame()

for i in file_list:
    data = pd.read_excel(path + i)
    y1 = pd.concat([y1,data])
y1 = y1.reset_index(drop=True)
y1
```

```python
y1.info()
```

```python
y1.sort_values("DateTime", inplace=True)
y1.reset_index(drop = True, inplace=True)
y1
```
