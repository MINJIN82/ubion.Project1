## IRS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
from scipy.stats import shapiro

irs = pd.read_excel("./엑셀/찐_IRS_10.xlsx", index_col=0)
irs
```

```python
irs = irs[["Mid"]]
irs
```

```python
irs["증감률"] =  (irs["Mid"] - irs["Mid"].shift(1)) / irs["Mid"]
irs.dropna(inplace=True)
irs.reset_index(drop=True, inplace=True)
irs
```

```python
sns.distplot(irs['증감률'])
```

```python
shapiro_test = stats.shapiro(irs["증감률"])
shapiro_test
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(irs['증감률'], plot=plt)
```

## CRS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
from scipy.stats import shapiro

crs = pd.read_excel("./엑셀/찐_CRS_10.xlsx", index_col=0)
crs = crs[["Mid"]]

crs["증감률"] =  (crs["Mid"] - crs["Mid"].shift(1)) / crs["Mid"]
crs.dropna(inplace=True)
crs.reset_index(drop=True, inplace=True)
crs
```

```python
crs["Mid"]
crs.loc[crs['Mid'] == 0]
```

```python
crs = crs[crs.Mid != 0]
```

```python
# crs = crs[crs.증감률 != 0]
# crs["증감률"] = crs["증감률"].round(4)
# crs
```

```python
def inf_to_nan(x):
    if x == float('inf') or x == float('-inf'):
        return float('nan')
    return float(x)
```

```python
crs["증감률"] = crs["증감률"].apply(inf_to_nan)
```

```python
sns.distplot(crs["증감률"])
```

```python
crs.info()
```

```python
stats.probplot(crs['증감률'], plot=plt)
```

## Basis Swap

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
from scipy.stats import shapiro

sb = pd.read_excel("./엑셀/SwapBasis_10.xlsx", index_col=0)
sb = sb[["DateTime", "1Y_베이시스"]]

sb["증감률"] =  (sb["1Y_베이시스"] - sb["1Y_베이시스"].shift(1)) / sb["1Y_베이시스"]
sb.rename({"1Y_베이시스" : "Basis_1Y"}, axis=1, inplace=True)
sb.dropna(inplace=True)
sb.reset_index(drop=True, inplace=True)
sb
```

```python
sb = sb[sb.Basis_1Y != 0]
sb.reset_index(drop=True, inplace=True)
sb
```

```python
sns.distplot(sb['증감률'])
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(sb['증감률'], plot=plt)
```
