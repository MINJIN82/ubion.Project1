```python
import pandas as pd
import numpy as np

# 데이터 끌고오기 및 삭제
df = pd.read_excel("../데이터자료/join_data.xlsx")
df = df.drop(['Unnamed: 0'], axis=1)
df.columns
```

```python
#스왑포인트 제외 파생컬럼(등락률) 생성

df["전일비_ex"] = (df["종가_ex"] - df["종가_ex"].shift(1)) / df["종가_ex"].shift(1)
df["전일비_1Y_irs"] = (df["1Y_Mid_irs"] - df["1Y_Mid_irs"].shift(1)) / df["1Y_Mid_irs"].shift(1)
df["전일비_2Y_irs"] = (df["2Y_Mid_irs"] - df["2Y_Mid_irs"].shift(1)) / df["2Y_Mid_irs"].shift(1)
df["전일비_3Y_irs"] = (df["3Y_Mid_irs"] - df["3Y_Mid_irs"].shift(1)) / df["3Y_Mid_irs"].shift(1)
df["전일비_5Y_irs"] = (df["5Y_Mid_irs"] - df["5Y_Mid_irs"].shift(1)) / df["5Y_Mid_irs"].shift(1)
df["전일비_10Y_irs"] = (df["10Y_Mid_irs"] - df["10Y_Mid_irs"].shift(1)) / df["10Y_Mid_irs"].shift(1)
df["전일비_1Y_crs"] = (df["1Y_Mid_crs"] - df["1Y_Mid_crs"].shift(1)) / df["1Y_Mid_crs"].shift(1)
df["전일비_2Y_crs"] = (df["2Y_Mid_crs"] - df["2Y_Mid_crs"].shift(1)) / df["2Y_Mid_crs"].shift(1)
df["전일비_3Y_crs"] = (df["3Y_Mid_crs"] - df["3Y_Mid_crs"].shift(1)) / df["3Y_Mid_crs"].shift(1)
df["전일비_5Y_crs"] = (df["5Y_Mid_crs"] - df["5Y_Mid_crs"].shift(1)) / df["5Y_Mid_crs"].shift(1)
df["전일비_10Y_crs"] = (df["10Y_Mid_crs"] - df["10Y_Mid_crs"].shift(1)) / df["1Y_Mid_crs"].shift(1)
df["전일비_국고1년"] = (df["국고1년"] - df["국고1년"].shift(1)) / df["국고1년"].shift(1)
df["전일비_국고3년"] = (df["국고3년"] - df["국고3년"].shift(1)) / df["국고3년"].shift(1)
df["전일비_국고5년"] = (df["국고5년"] - df["국고5년"].shift(1)) / df["국고5년"].shift(1)
df["전일비_국고10년"] = (df["국고10년"] - df["국고10년"].shift(1)) / df["국고10년"].shift(1)
df["전일비_통안364일"] = (df["통안364일"] - df["통안364일"].shift(1)) / df["통안364일"].shift(1)
df["전일비_통안2년"] = (df["통안2년"] - df["통안2년"].shift(1)) / df["통안2년"].shift(1)
df["전일비_1Y_베이시스"] = (df["1Y_베이시스"] - df["1Y_베이시스"].shift(1)) / df["1Y_베이시스"].shift(1)
df["전일비_2Y_베이시스"] = (df["2Y_베이시스"] - df["2Y_베이시스"].shift(1)) / df["2Y_베이시스"].shift(1)
df["전일비_3Y_베이시스"] = (df["3Y_베이시스"] - df["3Y_베이시스"].shift(1)) / df["3Y_베이시스"].shift(1)
df["전일비_5Y_베이시스"] = (df["5Y_베이시스"] - df["5Y_베이시스"].shift(1)) / df["5Y_베이시스"].shift(1)
df["전일비_10Y_베이시스"] = (df["10Y_베이시스"] - df["10Y_베이시스"].shift(1)) / df["10Y_베이시스"].shift(1)
df["전일비_ndf"] = (df["Mid_ndf"] - df["Mid_ndf"].shift(1)) / df["Mid_ndf"].shift(1)
df["M1_스왑포인트(원)"] = df["M1_스왑포인트"]/100 
df.head()
```

```python
# 등락률 중 분모에 0이들어가는 경우에 0이 2개연속이면 0 , 0이 전행에만 있는 경우이면 그 값을 그대로 넣어준다
for i in range(1, 2459):
    # "1Y_Mid_crs"
    if df.loc[i, "1Y_Mid_crs"] == 0 and df.loc[i-1,"1Y_Mid_crs"] == 0:
        df.loc[i,"전일비_1Y_crs"] = 0
    elif df.loc[i, "1Y_Mid_crs"] != 0 and df.loc[i-1,"1Y_Mid_crs"] == 0:
        df.loc[i,"전일비_1Y_crs"] = df.loc[i,"1Y_Mid_crs"]
        
    # "2Y_Mid_crs"  
    if df.loc[i, "2Y_Mid_crs"] == 0 and df.loc[i-1,"2Y_Mid_crs"] == 0:
        df.loc[i,"전일비_2Y_crs"] = 0
    elif df.loc[i, "2Y_Mid_crs"] != 0 and df.loc[i-1,"2Y_Mid_crs"] == 0:
        df.loc[i,"전일비_2Y_crs"] = df.loc[i,"2Y_Mid_crs"]
        
    # "3Y_Mid_crs" 
    if df.loc[i, "3Y_Mid_crs"] == 0 and df.loc[i-1,"3Y_Mid_crs"] == 0:
        df.loc[i,"전일비_3Y_crs"] = 0
    elif df.loc[i, "3Y_Mid_crs"] != 0 and df.loc[i-1,"3Y_Mid_crs"] == 0:
        df.loc[i,"전일비_3Y_crs"] = df.loc[i,"3Y_Mid_crs"]
        
    # "5Y_Mid_crs"   
    if df.loc[i, "5Y_Mid_crs"] == 0 and df.loc[i-1,"5Y_Mid_crs"] == 0:
        df.loc[i,"전일비_5Y_crs"] = 0
    elif df.loc[i, "5Y_Mid_crs"] != 0 and df.loc[i-1,"5Y_Mid_crs"] == 0:
        df.loc[i,"전일비_5Y_crs"] = df.loc[i,"5Y_Mid_crs"]
    
    # "10Y_Mid_crs"   
    if df.loc[i, "10Y_Mid_crs"] == 0 and df.loc[i-1,"10Y_Mid_crs"] == 0:
        df.loc[i,"전일비_10Y_crs"] = 0
    elif df.loc[i, "10Y_Mid_crs"] != 0 and df.loc[i-1,"10Y_Mid_crs"] == 0:
        df.loc[i,"전일비_10Y_crs"] = df.loc[i,"10Y_Mid_crs"]
        
    # "1Y_베이시스"   
    if df.loc[i, "1Y_베이시스"] == 0 and df.loc[i-1,"1Y_베이시스"] == 0:
        df.loc[i,"전일비_1Y_베이시스"] = 0
    elif df.loc[i, "1Y_베이시스"] != 0 and df.loc[i-1,"1Y_베이시스"] == 0:
        df.loc[i,"전일비_1Y_베이시스"] = df.loc[i,"1Y_베이시스"]
    
    # "2Y_베이시스"   
    if df.loc[i, "2Y_베이시스"] == 0 and df.loc[i-1,"2Y_베이시스"] == 0:
        df.loc[i,"전일비_2Y_베이시스"] = 0
    elif df.loc[i, "2Y_베이시스"] != 0 and df.loc[i-1,"2Y_베이시스"] == 0:
        df.loc[i,"전일비_2Y_베이시스"] = df.loc[i,"2Y_베이시스"]
    
    # "3Y_베이시스"            
    if df.loc[i, "3Y_베이시스"] == 0 and df.loc[i-1,"3Y_베이시스"] == 0:
        df.loc[i,"전일비_3Y_베이시스"] = 0
    elif df.loc[i, "3Y_베이시스"] != 0 and df.loc[i-1,"3Y_베이시스"] == 0:
        df.loc[i,"전일비_3Y_베이시스"] = df.loc[i,"3Y_베이시스"]            
     
    # "국고3년"    
    if df.loc[i, "국고3년"] == 0 and df.loc[i-1,"국고3년"] == 0:
        df.loc[i,"전일비_국고3년"] = 0
    elif df.loc[i, "국고3년"] != 0 and df.loc[i-1,"국고3년"] == 0:
        df.loc[i,"전일비_국고3년"] = df.loc[i,"국고3년"]
        
    # "통안364일"    
    if df.loc[i, "통안364일"] == 0 and df.loc[i-1,"통안364일"] == 0:
        df.loc[i,"전일비_통안364일"] = 0
    elif df.loc[i, "통안364일"] != 0 and df.loc[i-1,"통안364일"] == 0:
        df.loc[i,"전일비_통안364일"] = df.loc[i,"통안364일"]
    
    # "통안2년"    
    if df.loc[i, "통안2년"] == 0 and df.loc[i-1,"통안2년"] == 0:
        df.loc[i,"전일비_통안2년"] = 0
    elif df.loc[i, "통안2년"] != 0 and df.loc[i-1,"통안2년"] == 0:
        df.loc[i,"전일비_통안2년"] = df.loc[i,"통안2년"]
        
 
```
